### Box plot for finding outliers of credit amount

```python
#box plot to find outliers
sns.boxplot(y="Credit_amount",data=df1)
```

```python
#find IQR
Q1 = df1["Credit_amount"].quantile(0.25)
Q3 = df1["Credit_amount"].quantile(0.75)
IQR = Q3-Q1
Lower = Q1 - 1.5*IQR
Upper = Q3 + 1.5*IQR
```

```python
#Find outliers
Outliers = []
for i in range(len(df1["Credit_amount"])):
    if df1["Credit_amount"][i] < Lower or df1["Credit_amount"][i] > Upper:
        Outliers.append(i)
        
Outliers        
```
