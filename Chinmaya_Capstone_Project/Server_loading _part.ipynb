### Before load data into SQL server, first we have to normalize the dataset.
### To normalizing datasets we  have to perform normalization techniques like 1NF, 2NF, 3NF and BCNF.
### We have to speparte dataset into diffrent table to reduce redundancy and ensure that related data is stored in each table
### Also we have to create some unique columns to connect tables to each other. 

```python
# creating country_id  column with unique value for primary key
df2['Country_id'] = df2.index + 7000
df2
```

```python
# creating name_id  column with unique value for primary key
df2['name_id']=df2.index
df2
```

## To avoid transitive dependency we have to create tables accoring 3NF with containing a unique primary key. 

```python
# Separating data before load into SQl server
df4=df2[['Country_id','country']]
df4
```

```python
# creating Country_info table into SQL sever and load the data through engine
df4.to_sql('Country_info',engine,index=False)
```
